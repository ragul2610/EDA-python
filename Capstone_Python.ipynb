```python
#Seprating out the categorical and continuous variables-----> unique and nunique
from tabulate import tabulate
numeric_columns = []
categorical_columns = []

for column in tmdb.columns:
    if pd.api.types.is_numeric_dtype(tmdb[column]):
        numeric_columns.append(column)
    else:
        categorical_columns.append(column)
print(tabulate({"Categorical":categorical_columns,
                "numerical": numeric_columns}, headers = ["categorical", "numerical"]))
```

## * It is found that homepage,overview,release date,runtime and tagline have null values

```python
tmdb["runtime"].fillna(tmdb["runtime"].mean(),inplace=True)
```

```python
tmdb.isnull().sum()
```

# *Since imputing run time with mean will not affect data set and imputing other which are in Categorical will change the entire dataset.Replacing the null values of categorical as unknown

```python
def num(*s):
    for col in s:
        print(f"central tendencies of {col}: ")
        print(f"mean of {col} : {round(tmdb[col].mean(),2)}")
        print(f"median of {col} : {(tmdb[col].median())}")
        print(f"mode of {col} : {(tmdb[col].mode()[0])}")
        print()
```
